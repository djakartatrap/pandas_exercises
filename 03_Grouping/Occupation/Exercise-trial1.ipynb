### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep="|")
users.head(10)
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').mean()[['age']]
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
sum_frame = pd.concat([users, pd.get_dummies(users['gender'], prefix='gender')], axis=1).groupby('occupation').sum()
sum_frame['ratio_M'] = sum_frame.gender_M / (sum_frame.gender_F + sum_frame.gender_M)
sum_frame.sort_values('ratio_M', ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
maxmin_df = pd.concat([users.groupby('occupation')[['age']].max(),users.groupby('occupation')[['age']].min()], axis=1)
maxmin_df.columns = ['max', 'min']
maxmin_df
```

```python
users.groupby('occupation').age.agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).mean()[['age']]
```

### Step 8.  For each occupation present the percentage of women and men

```python
concat_df = pd.concat([users, pd.get_dummies(users['gender'], prefix='gender')], axis=1).groupby('occupation').sum()[['gender_F', 'gender_M']]
concat_df['ratio_M'] = concat_df['gender_M'] / (concat_df['gender_M'] + concat_df['gender_F'])
concat_df['ratio_F'] = concat_df['gender_F'] / (concat_df['gender_M'] + concat_df['gender_F'])
concat_df.loc[:, ['gender_M', 'gender_F', 'ratio_M', 'ratio_F']]
```

```python
# create a data frame and apply count to gender
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
```

```python
gender_ocup
```

```python
# create a DataFrame and apply count for each occupation
occup_count = users.groupby(['occupation']).agg('count')
```

```python
occup_count
```

```python
# divide the gender_ocup per the occup_count and multiply per 100
occup_gender = gender_ocup.div(occup_count, level = "occupation") * 100
```
